```python
labels = data['Class'].values
data=data.drop(['Class'], axis=1)
```

```python

from sklearn.discriminant_analysis import StandardScaler
scaler = StandardScaler()
df= scaler.fit_transform(data)
```

```python
from sklearn.ensemble import IsolationForest
iso_model = IsolationForest(random_state=42, contamination='auto')
iso_model=iso_model.fit(df)
data['Anomaly_iso'] = iso_model.predict(df)
print(data['Anomaly_iso'].value_counts())
data.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Convert prediction to same format as 'Class'
predicted_labels_iso = (data['Anomaly_iso'] == -1).astype(int)
print('\nIsolation Forest Evaluation:')
print(confusion_matrix(labels, predicted_labels_iso))
print(classification_report(labels, predicted_labels_iso, digits=4))
```

isolation forest model:
<li>High recall on  fraud(81.9%): Isolation Forest successfully detects most fraud cases
<li>Very Low Precision on Fraud(4.1%):The model has many false positives (normal transactions flagged as fraud).
<li>High Precision & Recall on Normal Transaction

```python
from sklearn.svm import OneClassSVM
oc_svm = OneClassSVM(kernel='rbf', nu=0.01, gamma='auto')
oc_svm.fit(df) 
data['Anomaly_svm'] = oc_svm.predict(df)
print(data['Anomaly_svm'].value_counts())
predicted_labels_svm = (data['Anomaly_svm'] == -1).astype(int)
print('\nSVM Evaluation:')
print(confusion_matrix(labels, predicted_labels_svm))
print(classification_report(labels, predicted_labels_svm, digits=4))
```

one class svm model:
<li>Low precision (8.9%):Out of all transactions SVM marked as fraud (2840), only 255 were truly fraud.
<li>Moderate recall (51.8%)Caught about half the actual frauds.
<li>High accuracy but misleading:99% sounds great, but it's mostly due to many normal transactions.


One-Class SVM performs worse than Isolation Forest 

```python
flagged_by_iso = data[data['Anomaly_iso'] == -1]
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(flagged_by_iso['Time'], bins=50, kde=True)
plt.title('Time Distribution of Isolation Forest Flagged Transactions')

plt.subplot(1, 2, 2)
sns.histplot(flagged_by_iso['Amount'], bins=50, kde=True)
plt.title('Amount Distribution of Isolation Forest Flagged Transactions')
plt.tight_layout()
plt.savefig('iso_forest_flagged_transactions_distribution.png')
plt.show()
```

<h2>Conclusion:</h2>

After applying the unsupervised models (Isolation Forest and One-Class SVM), The visualizations of "Time" and "Amount" distributions for these flagged transactions provide initial insights:

<li>Time Distribution of Flagged Transactions:Observe if the flagged transactions cluster around specific time periods. For instance, if fraudulent activities tend to occur during off-peak hours (e.g., late at night or early morning) or during specific days of the week, the time distribution of flagged transactions might show peaks in these unusual periods. This could indicate a pattern where fraudsters exploit times when monitoring might be less stringent.

<li>Amount Distribution of Flagged Transactions:Fraudulent transactions might involve unusually high amounts  or unusually low amounts . Deviations from the typical "Amount" distribution of legitimate transactions can be a strong indicator of anomaly.
